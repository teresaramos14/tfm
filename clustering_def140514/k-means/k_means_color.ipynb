```python
import pandas as pd
import numpy as np
import os
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
import umap
from yellowbrick.cluster import SilhouetteVisualizer 
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
folder_path = r"C:\Users\tere1\OneDrive\Escritorio\TFM\datos_img"
# Cargar las características normalizadas por Z-score
df_color_z = pd.read_csv(os.path.join(folder_path, "features_color_all_zscore.csv"))
# Cargar las características normalizadas por Min-Max
df_color_mm = pd.read_csv(os.path.join(folder_path, "features_color_all_minmax.csv"))
```

```python
df_color_z
```

```python
#Se eliminan los datos no numéricos: 'label' y 'filename'
df_color_z_cl= df_color_z.drop(columns=['filename', 'label'])
```

```python
# Método del codo
WCSS = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=1000, random_state=42)
    kmeans.fit(df_color_z_cl)
    WCSS.append(kmeans.inertia_)

# Graficar el codo
plt.plot(range(1, 11), WCSS, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('WCSS')
plt.title('Método del Codo para GLCM (z-score)')
plt.grid(True)
plt.show()
```

```python
# Creamos la figura con 2x2 subplots
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

# Probamos con k = 2, 3, 4, 5
for i, k in enumerate([2, 3, 4, 5]):
    km = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
    
    # Posición del subplot
    row = i // 2
    col = i % 2
    
    # Creamos el visualizador y ajustamos los datos
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[row][col])
    visualizer.fit(df_color_z_cl)  

plt.tight_layout()
plt.show()
```

```python
#Modelo k-means
cluster_color_z= KMeans(init='k-means++',n_clusters=2,random_state=42)
cluster_color_z.fit(df_color_z_cl)  
```

```python
#Se añaden las etiquetas y vemos en una tabla de contigencia cuantos smaller o higher hay en cada cluster
df_color_z['Cluster'] = cluster_color_z.labels_
pd.crosstab(df_color_z['Cluster'], df_color_z['label'])
```

```python
silhouette_avg_color_z = silhouette_score(df_color_z_cl, cluster_color_z.labels_)
print(f'Índice de silueta promedio: {silhouette_avg_color_z}')
```

Las imágenes etiquetadas como SMALLER están repartidas entre los clusteres 0 y 1.

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(df_color_z_cl)

# Visualización
plt.figure(figsize=(8, 5))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=df_color_z['Cluster'], palette='Set2')
plt.title('Clusters KMeans (t-SNE)')
plt.xlabel('t-SNE Componente 1')
plt.ylabel('t-SNE Componente 2')
plt.legend(title='Clúster')
plt.grid(True)
plt.show()
```

```python
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(df_color_z_cl)

# Visualización
plt.figure(figsize=(8, 5))
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=df_color_z['Cluster'], palette='Set2')
plt.title('Clusters KMeans (UMAP)')
plt.xlabel('UMAP Componente 1')
plt.ylabel('UMAP Componente 2')
plt.legend(title='Clúster')
plt.grid(True)
plt.show()
```

```python
# 1. Reducción de dimensionalidad con UMAP
umap_model = umap.UMAP(n_neighbors=15, n_components=5, min_dist=0.01, random_state=42)
X_umap = umap_model.fit_transform(df_color_z_cl)

# 2. Aplicar KMeans sobre el espacio reducido
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_umap)

# 3. Añadir los clústeres al dataframe original
df_color_z['Cluster_UMAP'] = clusters

# 4. Visualizar los resultados
plt.figure(figsize=(8, 6))
plt.scatter(
    X_umap[:, 0], X_umap[:, 1],
    c=clusters, cmap='Set1', edgecolor='k', alpha=0.7
)
plt.title('KMeans aplicado sobre datos reducidos con UMAP')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.colorbar(label='Clúster')
plt.grid(True)
plt.show()

# 5. Índice de silueta
silhouette = silhouette_score(X_umap, clusters)
print(f"Índice de silueta (UMAP + KMeans): {silhouette:.3f}")
```

```python
pd.crosstab(df_color_z['label'], df_color_z['Cluster_UMAP'])
```

```python
# 1. Reducción de dimensionalidad con UMAP
umap_model_mm = umap.UMAP(n_neighbors=10, n_components=5, min_dist=0.01, random_state=42)
X_umap_mm = umap_model_mm.fit_transform(df_color_mm_cl)

# 2. Aplicar KMeans sobre el espacio reducido
kmeans_mm = KMeans(n_clusters=4, init='k-means++', random_state=42)
clusters_mm = kmeans_mm.fit_predict(X_umap_mm)

# 3. Añadir los clústeres al dataframe original
df_color_mm['Cluster_UMAP'] = clusters_mm

# 4. Visualizar los resultados
plt.figure(figsize=(8, 6))
plt.scatter(
    X_umap_mm[:, 0], X_umap_mm[:, 1],
    c=clusters_mm, cmap='Set1', edgecolor='k', alpha=0.7
)
plt.title('KMeans aplicado sobre datos reducidos con UMAP')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.colorbar(label='Clúster')
plt.grid(True)
plt.show()

# 5. Índice de silueta
silhouette_mm = silhouette_score(X_umap_mm, clusters_mm)
print(f"Índice de silueta (UMAP + KMeans): {silhouette_mm:.3f}")
```

```python
pd.crosstab(df_color_mm['label'], df_color_mm['Cluster_UMAP'])
```

Se obtienen 4 clusteres muy equilibrados en número de imágenes smaller y higher. Puede que el color no permita separar clramente las dos clses. Por tanto, el color no sería un buen predictor de callos grandes y pequeños. Habrá cuatro agrupaciones distintas en los datos basadas en las características de color independientes de higher y smaller.
